```python
from pyspark.sql.functions import *

silver_df = spark.table(
    "zordering.silver.employee_salary_summary"
)

gold_dept_summary = (
    silver_df
    .groupBy("department_name")
    .agg(
        count("employee_id").alias("employee_count"),
        round(avg("base_salary"), 2).alias("avg_base_salary"),
        round(avg("bonus_amount"), 2).alias("avg_bonus_amount"),
        round(sum("bonus_amount"), 2).alias("total_bonus_paid"),
        round(sum("final_salary"), 2).alias("total_salary_paid")
    )
)

gold_dept_summary.write.mode("overwrite").saveAsTable(
    "zordering.gold.department_salary_summary"
)
```

```python
from pyspark.sql.functions import *

silver_df = spark.table(
    "zordering.silver.employee_salary_summary"
)

gold_monthly_summary = (
    silver_df
    .groupBy("month")
    .agg(
        countDistinct("employee_id").alias("employee_count"),
        round(sum("final_salary"), 2).alias("total_salary"),
        round(sum("bonus_amount"), 2).alias("total_bonus"),
        round(avg("final_salary"), 2).alias("avg_salary")
    )
)

gold_monthly_summary.write.mode("overwrite").saveAsTable(
    "zordering.gold.monthly_salary_summary"
)
```

```python
from pyspark.sql.functions import *

silver_df = spark.table(
    "zordering.silver.employee_salary_summary"
)

gold_kpi = (
    silver_df
    .groupBy("department_name")
    .agg(
        countDistinct("employee_id").alias("employees"),
        round(avg("final_salary"), 2).alias("avg_salary"),
        round(max("final_salary"), 2).alias("highest_salary"),
        round(min("final_salary"), 2).alias("lowest_salary"),
        round(sum("bonus_amount"), 2).alias("total_bonus")
    )
)

gold_kpi.write.mode("overwrite").saveAsTable(
    "zordering.gold.department_kpi"
)
```
