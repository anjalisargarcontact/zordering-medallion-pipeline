```python
%sql
CREATE SCHEMA IF NOT EXISTS zordering.bronze;
```

```python
# Read 3 CSV files into 3 DataFrames

# Read files

departments_df = spark.read.csv(
    "/Volumes/dep_catlog/default/dep_source/departments.csv",
    header=True
)

employee_df = spark.read.csv(
    "/Volumes/dep_catlog/default/dep_source/employee.csv",
    header=True
)

salaries_df = spark.read.csv(
    "/Volumes/dep_catlog/default/dep_source/salaries.csv",
    header=True
)

# Save as bronze tables

departments_df.write.mode("overwrite").saveAsTable("zordering.bronze.departments_raw")

employee_df.write.mode("overwrite").saveAsTable("zordering.bronze.employee_raw")

salaries_df.write.mode("overwrite").saveAsTable("zordering.bronze.salaries_raw")
```
